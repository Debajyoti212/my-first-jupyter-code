#FIND ALL THE RIDES FROM NEW YORK

```python
df1=df.loc[df["START*"]=="New York",["START*","STOP*"]]
df1.head()
```

#STARTING FROM NEWYORK ENDING IN QUEENS

```python
df.loc[(df["START*"] == "New York") & (df["STOP*"] == "Queens"), ["START*", "STOP*"]]
```

#STARTING FROM 2 POINTS

```python
#& is  not used because here and operator is not applicable or is applicable so we used|
df.loc[(df["START*"] == "Cary") | (df["START*"] == "New York")]
```
