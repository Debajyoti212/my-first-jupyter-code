#Conditionally adding columns

```python
import numpy as np
df["MILES REVIEW*"] = np.where(df["MILES*"] > 10, "long drive", np.where(df["MILES*"] < 10, "short drive", "NaN")) # Replace np.nan with a string "NaN"
                                                                                                                        # or any other string representation for missing value.
```
